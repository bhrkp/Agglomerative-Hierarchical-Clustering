```python
# Write your code for Task 1 here
random.seed(423)
Observation = []
for i in range(100):
  obs = []
  for j in range(3):
    obs.append(random.randint(-50,50))
  Observation.append(obs)

print(Observation)
```

```python
# Write your code for Task 2 here; you may use the following template
def distance(point1, point2):
    # write the content of this function
    # returns a number which is the Euclidean distance between the two points
    # Euclidean distance is the square root of (x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2
    #print('points (D):', point1,point2)
    sqrdis = sum(pow((p2 - p1),2) for p1,p2 in zip(point1,point2))
    return pow(sqrdis,0.5)
```

```python
# Write your code for Task 3 here; you may use the following template
class Node:
    def __init__(self, x):
        # write the content of the constructor here;
        # this constructor is used to initalize the single-point clusters in AHC
        self.cluster = x
        self.height = 0
        self.leftChild = None
        self.rightChild = None


    def getCluster(self):
      return self.cluster
    def getHeight(self):
      return self.height
    def getLeftChild(self):
      return self.leftChild
    def getRightChild(self):
      return self.rightChild

    def setHeight(self, h):
      self.height = h
    def setLeftChild(self, childNode):
      self.leftChild = childNode
    def setRightChild(self, childNode):
      self.rightChild = childNode
```

```python
currentClusters = []
for x in Observation:
  newNode = Node([x])
  currentClusters.append(newNode)


print(currentClusters)
```

```python
# Write your code for Task 4 here; you may use the following template
def singleLinkage(node1, node2):
    # write the content of this function
    # returns a number which is the single linkage between the two clusters in two nodes
    minDistance = float('inf')

    #print(node1,node2)

    cluster1 = node1.getCluster()
    cluster2 = node2.getCluster()

    #print('clusters (SL):', cluster1, cluster2, '\n')

    for point1 in cluster1:
      for point2 in cluster2:
        dis = distance(point1, point2)
        if dis < minDistance:
          minDistance = dis

    return minDistance


def completeLinkage(node1, node2):
    # complete linkage
    maxDistance = float('-inf')

    cluster1 = node1.getCluster()
    cluster2 = node2.getCluster()

    for point1 in cluster1:
      for point2 in cluster2:
        dis = distance(point1,point2)
        if dis > maxDistance:
          maxDistance = dis

    return maxDistance

def averageLinkage(node1, node2):
    # average linkage
    totalDistance = 0


    cluster1 = node1.getCluster()
    cluster2 = node2.getCluster()

    for point1 in cluster1:
      for point2 in cluster2:
        dis = distance(point1,point2)
        totalDistance += dis


    return totalDistance/(len(cluster1) + len(cluster2))

def centroidLinkage(node1, node2):
    # centroid linkage

    cluster1 = node1.getCluster()
    cluster2 = node2.getCluster()

    mean1 = 0
    mean2 = 0

    for point1 in cluster1:
      for point2 in cluster1:
        mean1 += distance(point1,point2)

    for point1 in cluster2:
      for point2 in cluster2:
        mean2 += distance(point1,point2)


    mean1 = mean1/len(cluster1)
    mean2 = mean2/len(cluster2)

    return abs(mean1-mean2)
```

```python
# Write your code for Task 5 here; you may use the following template
def clustersToMerge(currentClusters, linkage): # when this function is used,
    #   linkage should be one of the four linkage functions you implemented for Task 4
    # write the content of this function
    # returns a tuple (i,j,h) which indicates that the i-th and j-th nodes should be merged next
    #   to form a new Node with height h
  mindist = float('inf')
  mergeind = None

  for i in range(len(currentClusters) - 1):
    for j in range(i + 1, len(currentClusters)):
      #print(i,j)
      if linkage == 'single':
        dist = singleLinkage(currentClusters[i], currentClusters[j])
        #print('1')
      elif linkage == 'complete':
        dist = completeLinkage(currentClusters[i], currentClusters[j])
        #print('2')
      elif linkage == 'average':
        dist = averageLinkage(currentClusters[i], currentClusters[j])
        #print('3')
      elif linkage == 'centriod':
        dist = centroidLinkage(currentClusters[i], currentClusters[j])
        #print('4')
      #print('clusterToMerge', i, j, dist)
      if dist < mindist:
        #print(mergeind)
        mindist = dist
        mergeind = (i,j)
  if mergeind:
    if currentClusters[mergeind[0]].getHeight() >= currentClusters[mergeind[1]].getHeight():
      height = currentClusters[mergeind[0]].getHeight() + 1
    else:
      height = currentClusters[mergeind[1]].getHeight() + 1
    #print(mergeind, height)
    return (mergeind[0], mergeind[1], height)
  else: return None
```

```python
# Write your code for Task 6 here; you may use the following template
def ahc(observations, linkage):
    # write the content of this function
    # returns the root node of the AHC tree
    datapoints = observations.copy()
    while len(datapoints) > 1:

      i, j, height = clustersToMerge(datapoints, linkage)
      #print(i,j,height)
      node1Cluster = datapoints[i].getCluster()
      node2Cluster = datapoints[j].getCluster()
      cluster = node1Cluster + node2Cluster

      #print('cluster :', cluster)


      newnode = Node(cluster)
      newnode.setHeight(height)
      newnode.setRightChild(datapoints[i])
      newnode.setLeftChild(datapoints[j])
      #print(datapoints[i], datapoints[j])
      print('newnode:', newnode.getCluster(), newnode.getHeight(), '\n', newnode.getRightChild().getCluster(),newnode.getLeftChild().getCluster(), newnode.getHeight())
      datapoints.pop(j)
      datapoints.pop(i)
      datapoints.append(newnode)
      #print(newnode.getCluster())
      #print(datapoints)

    return datapoints[0]

```

```python

singleRoot = ahc(currentClusters,'single')
#x[0].getCluster()
print('Single: ','\n')

completeRoot = ahc(currentClusters,'complete')
#x[0].getCluster()
print('\n Complete: ', '\n')


averageRoot = ahc(currentClusters,'average')
#x[0].getCluster()
print('\n Average: ',' \n')

centriodRoot = ahc(currentClusters,'centriod')
#x[0].getCluster()
print('\n Centriod: ', '\n')
```

```python
# Write your code for Task 7 here; you may use the following template
def cutTree(rootNode, k):
  clusters = []
  stack = [rootNode]
  k -= 1

  while stack:
    if k == 0:
      break
    node = stack.pop(0)
    #print(node, node.getRightChild(), node.getLeftChild(), k)
    k += 1
    if node.getRightChild() is not None and k > 0:
      stack.append(node.getRightChild())
      k -= 1
    if node.getLeftChild() is not None and k > 0:
      stack.append(node.getLeftChild())
      k -=1
  for i in stack:
    clusters.append(i.getCluster())

  #print(len(stack))
  return clusters




linkage_roots = [singleRoot, completeRoot, averageRoot, centriodRoot]
linkage_name = ['single', 'complete', 'average', 'centriod']
for i in range(4):
  k = 5
  print('\n', linkage_name[i], ' :')
  clusters = cutTree(linkage_roots[i], k)
  #plot_clusters(currentClusters, clusters, linkage_name[i])
  for i in clusters:
    print(i)
```
